**FILTROS BASICOS PARA VARIABLES CUALITATIVAS**

```python
#Filtro por objeto 
#En este código filtro la columna NOMBRE_CLIENTE para que solo incluya el cliente: PÚBLICO EN GENERAL
filtro1=df[df["NOMBRE_CLIENTE"] == "PÚBLICO EN GENERAL"]
filtro1
```

```python
#Filtro por filas
#En este código filtro las filas 30, 60 y 90
#filtro2= df.iloc[3:10,: ]           #Filas consecutivas
filtro2=df.iloc[ [30, 60, 90] , : ]  #Filas no consecutivas 
filtro2
```

```python
#Filtro por columnas
#En este código filtro todas las filas con las columnas 1-3
filtro3=df.iloc[ : , 1:4]       #Columnas consecutivas 
#filtro3=df.iloc[ : , [1,4,6]]  #Columnas no consecutivas 
filtro3
```

```python
df1= pd.read_excel('Detalle precios y productos fabricados 2022.xlsx', index_col=2)
df1
```

```python
#Paso 2: Aplicamos Filtro de filas vs columnas 
#En este código filtro la información de la columna MARGEN_TOTAL_CALCULADO para el vendedor llamado: AARON NUÑO CORDOVA
filtro4=df1.loc[["AARON NUÑO CORDOVA"], ["MARGEN_TOTAL_CALCULADO"]]
filtro4
```

```python
#Filtro por cabecera
#En este código filtro todas las columnas para las 8 primeras instancias/filas
filtro5=df.head(8)
filtro5
```

**FILTROS BASICOS PARA VARIABLES CUANTITATIVAS**

```python
#Filtro por comparacion
#En este código filtro las filas cuyo valor en la columna MARGEN_TOTAL_CALCULADO es mayor a 50,000
filtro6=df[df["MARGEN_TOTAL_CALCULADO"] > 50000]
filtro6
```

**FILTROS BASICOS CON CONETORES LOGICOS**

```python
#Filtro Y
#En este código filtro las filas cuyo valor en la columna MARGEN_TOTAL_CALCULADO es menor a 10,000 y el vendedor de la columna NOMBRE_VENDEDOR es ALFREDO CANELA
filtro7=df[(df["MARGEN_TOTAL_CALCULADO"] < 10000) & (df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro7
```

```python
#Filtro o
#En este código filtro las filas cuyo valor en la columna MARGEN_TOTAL_CALCULADO es menor a 8,000 o el vendedor de la columna NOMBRE_VENDEDOR es ALFREDO CANELA
filtro8=df[(df["MARGEN_TOTAL_CALCULADO"] < 8000) | (df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro8
```

```python
#Filtro not
#En este código filtro las filas cuyo valor en la columna MARGEN_TOTAL_CALCULADO es menor a 8,000 y el vendedor de la columna NOMBRE_VENDEDOR no es ALFREDO CANELA
filtro9=df[~(df["MARGEN_TOTAL_CALCULADO"] >= 8000) & ~(df["NOMBRE_VENDEDOR"] == "ALFREDO CANELA")]
filtro9
```
