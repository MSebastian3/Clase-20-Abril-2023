```python
#Carga desde un archivo .xlsx sin indice
df= pd.read_excel('Datos de Facturación.xlsx')
#df
```

```python
#Verificamos información del DataFrame
df.info()
```

```python
#Imprimo los primeros 5 registro del dataframe
df.head()
```

```python
#1-CVE_CLPV
filtro1= df.iloc[1000:2000,: ]           #Filas consecutivas
#filtro2=df.iloc[ [30, 60, 90] , : ]  #Filas no consecutivas 
filtro1
```

```python
#2-CVE_VEND
#filtro2=df[~df['CVE_VEND'].isin([5.0, 4.0])]
filtro2=df[~(df["CVE_VEND"] == 5) & ~(df["CVE_VEND"] == 4)]
filtro2
```

```python
#3-FECHA_ENT
filtro3=df[df["FECHA_ENT"] == "2022-02-28 00:00:00"]
filtro3
```

```python
#4-CAN_TOT
filtro4=df[(df["CAN_TOT"] < 5951.7) | (df["STATUS"] == "E")]
filtro4
```

```python
#5- SOLO COLUMNAS CVE_DOC, FECHA_ENT, FECHA_VEN Y CAN_TOT
#filtro3=df.iloc[ : , 1:4]       #Columnas consecutivas 
filtro5=df.iloc[ : , [0,6,7,9]]  #Columnas no consecutivas 
filtro5
```

```python
#6- SOLO LAS FILAS 7001-7099
filtro6= df.iloc[7001:7100,: ]           #Filas consecutivas
filtro6
```

```python
#7- INDEX = CVE_VEND(valores: 1, 2) (columna: FECHAELAB)
df2 = df.set_index('CVE_VEND')
filtro7 = df2.loc[[1,2], ["FECHAELAB"]]
filtro7
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Entregable2.1.csv")
filtro2.to_csv("Entregable2.2.csv")
filtro3.to_csv("Entregable2.3.csv")
filtro4.to_csv("Entregable2.4.csv")
filtro5.to_csv("Entregable2.5.csv")
filtro6.to_csv("Entregable2.6.csv")
filtro7.to_csv("Entregable2.7.csv")
```
